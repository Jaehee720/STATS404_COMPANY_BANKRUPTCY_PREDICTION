```python
import pandas as pd
```

```python
file_name = "/Users/jeongjaehui/Documents/Stats_404/JEONG-JAEHEE/used_car_price_predict_traindata.csv"
df = pd.read_csv(filepath_or_buffer=file_name, encoding='latin-1', nrows = 6019)
```

```python
df = df.iloc[:, 1:]
df.columns
```

```python

df.columns
df.head

pd.options.display.max_columns = 50
df.tail()
```

```python
df['Year'].value_counts(sort=False)
from collections import Counter
Counter(df['Year'])
```

```python
#check names of cars
df_1 = df['Name']
df_1 = pd.DataFrame(df_1, columns = ['Name'])

#split string by space
df_2 = pd.DataFrame()
df_2 = df_1['Name'].str.split(" ", n=9, expand = True)

#check missing values
df_2.info()

#check car brands
Counter(df_2[0])

df_a = df
df_a['Brand'] = df_2[0]
df_a.head
```

```python
df_a.describe()
```

```python
df_a.info()
```

```python
#Some values of the Engine, Power, Seats variables are missing
#Most of the values of New_Price are missing. It's because this company is a selling used car company, they don't have many new cars.

df_a.fillna('Missing')
```

```python
%matplotlib inline
```

```python
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
import seaborn as sns

#Count Price to check distribution
sns.displot(df_a['Price'])
plt.xlim(0,160)

#relation between Brand and Price
plt.figure(figsize = (20, 10))
sns.catplot(x='Brand', kind='count', palette='ch:.25', height = 8, aspect = 2, data = df_a);
plt.xticks(rotation=90);

#relation between Year and Price
df_3 = pd.concat([df_a['Price'], df_a['Year']], axis=1)
f, ax = plt.subplots(figsize=(20, 10))
fig = sns.boxplot(x='Year', y="Price", data=df_3)
fig.axis(ymin=0, ymax=165);
plt.xticks(rotation=90);

#relation between Fuel_Type and Price
df_3 = pd.concat([df_a['Price'], df_a['Fuel_Type']], axis=1)
f, ax = plt.subplots(figsize=(12, 8))
fig = sns.boxplot(x='Fuel_Type', y="Price", data=df_3)
fig.axis(ymin=0, ymax=165);

#relation between Owner_ship and Price
df_3 = pd.concat([df_a['Price'], df_a['Owner_Type']], axis=1)
f, ax = plt.subplots(figsize=(12, 8))
fig = sns.boxplot(x='Owner_Type', y="Price", data=df_3)
fig.axis(ymin=0, ymax=165);

#check correlations
plt.figure(figsize = (5, 5))
sns.heatmap(df_a.corr(), annot = True)
plt.show()
```
