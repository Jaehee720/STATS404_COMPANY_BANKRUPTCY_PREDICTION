```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
#create connection to database file
database = 'chinook.db'
connection = sql.connect(database)
```

```python
#There are ties and it shows all the results who are the top 3 by revenue just like Olympic Games.

query ='''  select * from(
                select strftime('%Y', i.invoicedate) as year, sum(ii.unitprice * ii.quantity) as revenue, ar.name, ar.artistid,
                    rank () over (
                        partition by strftime('%Y', i.invoicedate)
                        order by sum(ii.unitprice * ii.quantity) desc
                    ) rank
                from invoices i
                left join invoice_items ii
                    on i.invoiceid = ii.invoiceid
                inner join tracks t
                    on ii.trackid = t.trackid
                inner join albums al
                    on t.albumid = al.albumid
                inner join artists ar
                    on al.artistid = ar.artistid
                  where billingstate = 'CA'
                  group by strftime('%Y', i.invoicedate), ar.name, ar.artistid
            )
            where rank <4
            '''
df = pd.read_sql_query(query, connection)
df
```
